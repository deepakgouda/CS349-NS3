```python
from pandas import read_csv
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
```

```python
versions = ['NewReno', 'Hybla', 'Westwood', 'Scalable', 'Vegas']
```

```python
for version in versions:
    data = read_csv('./Output/Tcp'+version+'.cwnd', delimiter='\s', header=None, engine='python')
    data.head()
    
    X = data[0]
    Y = data[2]
#     if version is 'NewReno':
#         Y = Y+1000
    plt.plot(X, Y, label=version)
plt.legend()
plt.show()
```

```python
for version in versions:
    data = read_csv('./Output/Tcp'+version+'-packet-byte-count-0.txt', delimiter='\t', header=None, engine='python')

    X = data[0]
    Y = np.cumsum(data[1])
    plt.plot(X, Y, label=version)
plt.legend()
plt.show()
```

```python
for version in versions:
    data = read_csv('./Output/Tcp'+version+'-packet-byte-count-1.txt', delimiter='\t', header=None, engine='python')

    X = data[0]
    Y = np.cumsum(data[1])
    plt.plot(X, Y, label=version)
plt.legend()
plt.show()
```

```python
for version in versions:
    data = read_csv('./Output/Tcp'+version+'.drop', delimiter='\t', header=None, engine='python')
    data[1] = 1
    X = data[0]
    Y = np.cumsum(data[1])
    plt.plot(X, Y, label=version)
plt.legend()
plt.show()
```
